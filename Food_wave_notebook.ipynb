```python
fig, ax = plt.subplots(figsize=[12,8])
sns.countplot(data=food_df_num, x=food_df_num['fd_sec_level'], ax=ax)
handltext = ["0", "2", "4", "8"]
labels = ["High food security", "Marginal food security", "Low food security", "Very low food security"]

from matplotlib.legend_handler import HandlerBase
from matplotlib.text import Text

font = {'size': 14,'weight': 'heavy'}

class TextHandler(HandlerBase):
    def create_artists(self, legend, tup ,xdescent, ydescent,
                        width, height, fontsize,trans):
        tx = Text(width/2.,height/2,tup[0], fontsize=fontsize,
                  ha="center", va="center", color=tup[1], fontweight="bold")
        return [tx]

n_points = food_df_num.shape[0]
max_count = food_df_num['fd_sec_level'].value_counts().max()
max_prop = max_count / n_points

tick_props = np.arange(0, max_prop + 0.05, 0.05)
tick_names = [f'{v:0.2f}' for v in tick_props]

t = ax.get_xticklabels()
labeldic = dict(zip(handltext, labels))
labels = [labeldic[h.get_text()]  for h in t]
handles = [(h.get_text(),c.get_fc()) for h,c in zip(t,ax.patches)]

ax.legend(handles, labels, handler_map={tuple : TextHandler()})
ax.set_yticks(tick_props * n_points)
ax.set_yticklabels(tick_names, fontsize=12)
ax.set_xlabel('food security level', fontdict=font)
ax.set_ylabel('Proportion', fontdict=font)
ax.text(-0.05, 2350, '0.8', color='w', size=14)
ax.text(0.95, 205, '0.1', color='w', size=14)
ax.text(1.935, 75, '0.06', color='w', size=14)
ax.text(2.925, 60, '0.04', color='w', size=14)

plt.show();
```
